# Yolo3 model export (can't convert to mxnet)

```python
def Yolov3_onnx():
    img = Variable(torch.randn(1, 3, 608, 608, ))
    dummy_input = torch.cat([img]) 
    
    ## model
    from yolo.darknet import Darknet
    det_model = Darknet("yolo/cfg/yolov3-spp.cfg") 
    det_model.load_weights('models/yolo/yolov3-spp.weights')
    det_model.net_info['height'] = 608
    det_model.cuda()
    det_model.eval()

    
    dummy_input = dummy_input.to('cuda')
    
    dummy_input.shape

    # visualize model
    y = det_model(dummy_input) 
#     make_dot(y.mean(), params=dict(det_model.named_parameters()z))
    
    # export onnx
    torch.onnx.export(det_model, dummy_input, "models/onnx/yolo.onnx")
Yolov3_onnx()
```
